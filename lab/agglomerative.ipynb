```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
%matplotlib inline
import pandas as pd
import matplotlib
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as shc
from scipy.cluster.hierarchy import fcluster
```

```python
df = pd.read_csv('Wholesale.csv')
#df1 = df.copy(deep=True)
df=df.drop(['Channel','Region'],axis=1)
```

```python
linked = shc.linkage(df, method='single')
```

```python
print (linked)
```

```python
plt.figure(figsize=(26, 26))  
dendrogram(linked)
plt.title("test")
plt.show()
#print('Optimim number of clusters :',8)
```

```python
hc = AgglomerativeClustering(n_clusters=8, affinity = 'euclidean', linkage = 'single')
```

```python
y_hc = hc.fit_predict(df)
```

```python
print(y_hc)
```

```python
df['predict'] = y_hc
print(df)
```

```python
linked1 = linkage(df, 'complete')
plt.figure(figsize=(26, 26))  
dendrogram(linked1)
plt.title("test")
plt.show()
```

```python
linked2 = linkage(df, 'average')
plt.figure(figsize=(26, 26))  
dendrogram(linked2)
plt.title("test")
plt.show()
```
