```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from sklearn import datasets
import math
```

```python
df = pd.read_csv('Wholesale.csv')
df1 = df.copy(deep=True)
df=df.drop(['Channel','Region'],axis=1)
```

```python
sum0=0
sum1=0
sum2=0
sum3=0
sum4=0
```

```python
s0=[]
s1=[]
s2=[]
s3=[]
s4=[]
l1=0
```

```python
# fit_transform function calculates the distance of each point from all the centroids. We then assign these points to nearest centroid. 
km1 = KMeans(n_clusters=5, n_init=25, max_iter = 600, random_state=0)
ald = km1.fit_transform(df)
for i in range (0,440):
    min = ald[i,0]
    for j in range(0,5):
        if ald[i,j]<=min:
            min = ald[i,j]
            k=j
    print(k,l1,'-->>')
    if k==0:
        sum0=sum0+min
        s0.append(min)
    elif k==1:
        sum1=sum1+min
        s1.append(min)
    elif k==2:
        sum2=sum2+min
        s2.append(min)
    elif k==3:
        sum3=sum3+min
        s3.append(min)
    elif k==4:
        sum4=sum4+min
        s4.append(min)
    l1=l1+1
l0=len(s0)
l1=len(s1)
l2=len(s2)
l3=len(s3)
l4=len(s4)
print(l0)
#print(sum0,sum1,sum2,sum3,sum4)
```

```python
df_centroids =  km1.labels_
df_centers = km1.cluster_centers_
```

```python
print(df_centers) #Generating Coordinates of the centroids of six clusters.
```

```python
#Defining which points from our datasets belong to which cluster.
for i in range(km1.n_clusters):
    i= np.where(km1.labels_ == i)[0]
    print (i)
```

```python
x0=0
x1=0
x2=0
x3=0
x4=0
```

```python
#Calculating no. of points in each cluster.
for i in range(0,len(df_centroids)):
    if df_centroids[i]==0:
        x0=x0+1
    if df_centroids[i] == 1:
        x1 = x1 + 1
    if df_centroids[i]==2:
        x2=x2+1
    if df_centroids[i]==3:
        x3=x3+1
    if df_centroids[i]==4:
        x4=x4+1
print(x0,x1,x2,x3,x4)
```

```python
#Method to calculate mean from each cluster to help with calculation of SSE within cluster
sum0_avg = sum0/x0
sum1_avg = sum1/x1
sum2_avg = sum2/x2
sum3_avg = sum3/x3
sum4_avg = sum4/x4
print(sum0_avg,sum1_avg,sum2_avg,sum3_avg,sum4_avg)
```

```python
# Within CLuster SSE
s0_0 = []
for i in range (0,l0):
    s0_0.append((s0[i]-sum0_avg)**2)
print('Within Cluster SSE of Cluster0',sum(s0_0))
s1_1=[]
for i in range (0,l1):
    s1_1.append((s1[i]-sum1_avg)**2)
print('Within Cluster SSE of Cluster1',sum(s1_1))
s2_2=[]
for i in range (0,l2):
    s2_2.append((s2[i]-sum2_avg)**2)
print('Within Cluster SSE of Cluster2',sum(s2_2))
s3_3=[]
for i in range (0,l3):
    s3_3.append((s3[i]-sum3_avg)**2)
print('Within Cluster SSE of Cluster3',sum(s3_3))
s4_4=[]
for i in range (0,l4):
    s4_4.append((s4[i]-sum4_avg)**2)
print('Within Cluster SSE of Cluster4',sum(s4_4))
```

```python
sse=[]
for i in range(1,10):
    km11=KMeans(n_clusters=i, n_init=25, max_iter = 600, random_state=0)
    km11.fit(df)
    sse.append(km11.inertia_)
print (sse) #Sum of squared distances of samples from their nearest cluster center with continuously increasing the number of clusters.
```

```python
km=KMeans(n_clusters=10, n_init=25, max_iter = 600, random_state=0)
km.fit(df)
print('Between Clusters SSE',km.inertia_)
```

```python
k = range(1,10)
plt.figure()
plt.plot(k,sse)
plt.xlabel('Number of Clusters')
plt.ylabel('Sse')
plt.show()
```
